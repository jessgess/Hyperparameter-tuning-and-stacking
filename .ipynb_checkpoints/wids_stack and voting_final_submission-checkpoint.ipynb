```python
import numpy as np 
import pandas as pd 
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier
import xgboost as xgb
from sklearn import svm
from sklearn.ensemble import VotingClassifier
from sklearn.cross_validation import KFold
import warnings
import matplotlib.pyplot as plt
# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

from subprocess import check_output
print(check_output(["ls", "../input"]).decode("utf8"))
pd.set_option('display.max_columns', 400) 
# Any results you write to the current directory are saved as output.
    
```

```python
data = pd.read_csv('../input/train.csv')
print(data.shape)
data.head()
```

```python
target = data['is_female'] 
data.drop(['train_id'], axis=1, inplace=True)
```

**Separating columns with more than 12K NaN rows:**

```python
high_nulls_1 = []
high_null_cols = []
for col in data.columns:
    null_sum = data[col].isnull().sum()
    if null_sum > 12000:
        high_nulls_1.append((col, null_sum))

for item in high_nulls_1:
    high_null_cols.append(item[0])
len(high_null_cols)
```

**Separating columns with quantity of NaN values between 12 & 17[](http://)K:**

```python
high_nulls_2 = []
for col in high_null_cols:
    null_sum = data[col].isnull().sum()
    if 12000 <= null_sum < 14000:
        high_nulls_2.append(col)
```

```python
data_hi_nan = data[high_nulls_2]
data_hi_nan.fillna(0, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  
data_hi_nan = pd.DataFrame(scaler.fit_transform(data_hi_nan),
                           columns = data_hi_nan.columns,
                           index = data_hi_nan.index )
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95) #0.95 generates 78 components hence 78 below
data_hi_nan_pca = pd.DataFrame(pca.fit_transform(data_hi_nan), 
                        columns = ['PC'+str(i+1) for i in range(10)], 
                        index   = data_hi_nan.index)

print(data_hi_nan_pca.shape)
```

** Columns with less than 12K Nan values**

```python
data_lo_nan = data.drop(high_null_cols, axis=1)
data_lo_nan.shape
```

```python
#labling the categorical data
le = preprocessing.LabelEncoder()
data_lo_nan['LN2_WIndLngBEOth'] = le.fit_transform(data_lo_nan['LN2_WIndLngBEOth'].fillna('NA')) 
data_lo_nan['LN2_RIndLngBEOth'] = le.fit_transform(data_lo_nan['LN2_RIndLngBEOth'].fillna('NA')) 
```

```python
data_lo_nan.drop('is_female', axis=1, inplace=True)
data_lo_nan.fillna(-99, inplace=True)
data_lo_nan.shape
```

**Including data_hi_nan_pca did not improve the results so reverting back to only using columns with less than 12K nan values (data_lo_nan)**

```python
data_concat = data_lo_nan.join(data_hi_nan_pca)
data_concat.shape
```

**Developing the models and validating them:**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_lo_nan, target,
                                                    random_state=19,
                                                    test_size=0.2,
                                                    stratify=target)
```

```python
#utility function for reporting model performance
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python
#RandomForest
param_grid = {"max_depth": [10, 12],
              "max_features": [300, 380],
              "min_samples_split": [3, 5],
              "min_samples_leaf": [3, 5]}

grid_search = GridSearchCV(RandomForestClassifier(n_estimators=75), param_grid=param_grid, cv=3)
grid_search.fit(X_train, y_train)
print("GridSearchCV for %d candidate parameter settings." % (len(grid_search.cv_results_['params'])))
report(grid_search.cv_results_)
```

```python
#best results: 0.9056 (0.004 std)
#changing max_features to 320 based on feature importances
rf = RandomForestClassifier(max_depth=14, max_features=320, min_samples_leaf=5,
                            min_samples_split=3, n_estimators=75, random_state=19)
rf.fit(X_train, y_train)
rf_score = cross_val_score(rf, X_train, y_train, cv=5, scoring='accuracy')
print('Avg cross validated score for RandomForest: %.4f (%.3f std)'% (rf_score.mean(), rf_score.std()))
```

```python
print("RF accuracy on test is:", accuracy_score(y_test, rf.predict(X_test)))
cm = confusion_matrix(y_test, rf.predict(X_test))
sns.heatmap(cm, annot=True, fmt="d")
```

**Making a 2-layer estimator by incorporate above estimators and predicting the actual test set:**

```python
#preparing the test dataset
train_cols = list(data_lo_nan.columns.values)
test = pd.read_csv('../input/test.csv')
test = test[train_cols]
test['LN2_WIndLngBEOth'] = le.fit_transform(test['LN2_WIndLngBEOth'].fillna('NA')) 
test['LN2_RIndLngBEOth'] = le.fit_transform(test['LN2_RIndLngBEOth'].fillna('NA'))
test.fillna(-99, inplace=True)
```

**Credits**: Code for below section on stacking is the work of Anisotropic, taken from: https://www.kaggle.com/arthurtok/introduction-to-ensembling-stacking-in-python?scriptVersionId=2150743

```python
train = data_lo_nan

# training and train/test split parameters
ntrain = train.shape[0]
ntest = test.shape[0]
SEED = 0 # for reproducibility
NFOLDS = 5 # set folds for out-of-fold prediction
kf = KFold(ntrain, n_folds= NFOLDS, random_state=SEED)

# Class to extend the Sklearn classifier
class SklearnHelper(object):
    def __init__(self, clf, seed=0, params=None):
        params['random_state'] = seed
        self.clf = clf(**params)

    def train(self, x_train, y_train):
        self.clf.fit(x_train, y_train)

    def predict(self, x):
        return self.clf.predict(x)
    
    def fit(self,x,y):
        return self.clf.fit(x,y)
```

```python
def get_oof(clf, x_train, y_train, x_test):
    oof_train = np.zeros((ntrain,))
    oof_test = np.zeros((ntest,))
    oof_test_skf = np.empty((NFOLDS, ntest))

    for i, (train_index, test_index) in enumerate(kf):
        x_tr = x_train[train_index]
        y_tr = y_train[train_index]
        x_te = x_train[test_index]

        clf.train(x_tr, y_tr)

        oof_train[test_index] = clf.predict(x_te)
        oof_test_skf[i, :] = clf.predict(x_test)

    oof_test[:] = oof_test_skf.mean(axis=0)
    return oof_train.reshape(-1, 1), oof_test.reshape(-1, 1)
```

```python
rf_params = {'max_depth':14, 
             'max_features':320,
             'min_samples_leaf':5,
             'min_samples_split':3,
             'n_estimators':75,
             'n_jobs':-1}
xgb_params = {'learning_rate':0.1,
              'max_depth':5,
              'min_child_weight':5,
              'n_estimators':300,
              'n_jobs':-1}
ext_params = {'max_features':320,
              'n_estimators':75,
              'max_depth':10,
              'class_weight':"balanced",
              'n_jobs':-1}
ada_params = {'n_estimators':400,
              'learning_rate':0.5}
'''knn_params = {'n_neighbors':10,
              'weights':'distance',
              'n_jobs':-1}'''
```

```python
# Create 5 objects that represent our 4 models
rf = SklearnHelper(clf=RandomForestClassifier, seed=SEED, params=rf_params)
gb = SklearnHelper(clf=xgb.XGBClassifier, seed=SEED, params=xgb_params)
et = SklearnHelper(clf=ExtraTreesClassifier, seed=SEED, params=ext_params)
ada = SklearnHelper(clf=AdaBoostClassifier, seed=SEED, params=ada_params)
#knn = SklearnHelper(clf=KNeighborsClassifier, seed=SEED, params=knn_params)
```

```python
y_train = target.ravel() #flattening the target 
x_train = train.values 
x_test = test.values
```

**Output of the First level Predictions**

Feeding the training and test data into our 5 base classifiers and use the Out-of-Fold prediction function we defined earlier to generate our first level predictions.

```python
# Create our OOF train and test predictions. These base results will be used as new features
rf_oof_train, rf_oof_test = get_oof(rf,x_train, y_train, x_test) # Random Forest
gb_oof_train, gb_oof_test = get_oof(gb,x_train, y_train, x_test) # Gradient Boost
et_oof_train, et_oof_test = get_oof(et, x_train, y_train, x_test) # Extra Trees
ada_oof_train, ada_oof_test = get_oof(ada, x_train, y_train, x_test) # AdaBoost 

print("Training is complete")
```

creating the heatmap of the 4 classifiers and check how correlated they are (the less the better):

```python
base_predictions_train = pd.DataFrame( {'RandomForest': rf_oof_train.ravel(),
                                        'GradientBoost': gb_oof_train.ravel(),
                                        'ExtraTrees': et_oof_train.ravel(),
                                        'AdaBoost': ada_oof_train.ravel()})
base_predictions_train.head()
```

```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
data = [go.Heatmap(z= base_predictions_train.astype(float).corr().values,
                   x=base_predictions_train.columns.values,
                   y= base_predictions_train.columns.values,
                   colorscale='Viridis', 
                   showscale=True, reversescale = True)]
py.iplot(data, filename='labelled-heatmap')
```

**Second-Level Predictions from the First-level Output:**

```python
x_train = np.concatenate((et_oof_train, rf_oof_train, ada_oof_train, gb_oof_train), axis=1)
x_test = np.concatenate((et_oof_test, rf_oof_test, ada_oof_test, gb_oof_test), axis=1)
```

**Second level learning model via XGBoost**

```python
#best params: {'gamma': 0.9, 'learning_rate': 0.1, 'max_depth': 4, 'min_child_weight': 2, 'n_estimators': 100} 
#accuracy = 0.914105724459
params={'n_estimators':[100, 300, 400],
        'max_depth':[2, 4, 6],
        'min_child_weight':[2, 4, 5],
        'learning_rate':[0.01, 0.1, 0.2],
        'gamma':[0.9, 1]}
stack_gb = GridSearchCV(xgb.XGBClassifier(), params,
                        scoring='accuracy', cv=5, n_jobs=-1)
stack_gb.fit(x_train, y_train)
print(stack_gb.best_params_, stack_gb.best_score_)
```

```python
stack = xgb.XGBClassifier(
    learning_rate=0.1,
    n_estimators=100,
    max_depth=4,
    min_child_weight=2,
    gamma=0.9,                        
    objective='binary:logistic',
    n_job=-1)
stack.fit(x_train, y_train)
predictions = stack.predict(x_test)
```

```python
#Mean score = 0.914, Std deviation = 0.005
scores = cross_val_score(stack, x_train, y_train, cv=5)
print(scores)
print("Mean score = %.3f, Std deviation = %.3f"%(np.mean(scores),np.std(scores)))
```
